a.Get the items not common to both series A and series B (Without
using loops)

```
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
```

```
union = pd.Series(np.union1d(ser1, ser2)) 

intersect = pd.Series(np.intersect1d(ser1, ser2)) 
   
notcommon = union[~union.isin(intersect)] 
  
# displaying the result 
print(notcommon) 
```

b.Get the items common to both series A and series B(Without using
loops)

```
union1 = pd.Series(np.union1d(ser1, ser2)) 

intersect1 = pd.Series(np.intersect1d(ser1, ser2)) 
   
common = union1[union.isin(intersect1)] 
  
# displaying the result 
print(common) 
```

5.Find the desired output as shown in the image below.

```
Train_data = pd.DataFrame({'State': ['A','B','C','A','A','B','A','A','C','C'],
'class': [0,1,1,0,1,1,0,1,1,0]},dtype=int)
Train_data
```

```
td= pd.crosstab(Train_data['State'],Train_data['class']).reset_index()
td
```

```
encoded_TrainData=pd.DataFrame(columns=['State_0','State_1'])
```

```
def encoding0(x):
  if x["State"] == 'A':
    y = str(td[td.State == 'A'][0].values[0])
    z = str(td[td.State == 'A'].sum(axis=1).values[0])
    return y + "/" + z
  elif x["State"] == 'B':
    y = str(td[td.State == 'B'][0].values[0])
    z = str(td[td.State == 'B'].sum(axis=1).values[0])
    return y + "/" + z
  elif x["State"] == 'C':
    y = str(td[td.State == 'C'][0].values[0])
    z = str(td[td.State == 'C'].sum(axis=1).values[0])
    return y + "/" + z
```

```
def encoding1(x):
  if x["State"] == 'A':
    y = str(td[td.State == 'A'][1].values[0])
    z = str(td[td.State == 'A'].sum(axis=1).values[0])
    return y + "/" + z
  elif x["State"] == 'B':
    y = str(td[td.State == 'B'][1].values[0])
    z = str(td[td.State == 'B'].sum(axis=1).values[0])
    return y + "/" + z
  elif x["State"] == 'C':
    y = str(td[td.State == 'C'][1].values[0])
    z = str(td[td.State == 'C'].sum(axis=1).values[0])
    return y + "/" + z
```

```
encoded_TrainData['State_0'] = Train_data.apply(lambda Class_value:encoding0(Class_value), axis=1)
encoded_TrainData['State_1'] = Train_data.apply(lambda Class_value:encoding1(Class_value), axis=1)
```

```
encoded_TrainData= encoded_TrainData[['State_0','State_1']]
encoded_TrainData
```
